```python
#-*- coding:utf-8 -*-
import re
import jieba
import os

def get_File_List(filePath):
    filenames = os.listdir(filePath)
    return filenames

#分别获得正常邮件、垃圾邮件及测试文件名称列表
normFileList = get_File_List(r"F:\NLP_learning\BayesSpam-master\data\normal")
spamFileList = get_File_List(r"F:\NLP_learning\BayesSpam-master\data\spam")
testFileList = get_File_List(r"F:\NLP_learning\BayesSpam-master\data\test")

#获取训练集中正常邮件与垃圾邮件的数量
normFilelen = len(normFileList)
spamFilelen = len(spamFileList)
print(normFilelen, spamFilelen)

#获取停用词列表
def getStopWords():
    stopList = []
    for line in open( r"F:\NLP_learning\BayesSpam-master\data\中文停用词表.txt"):
        stopList.append(line[:len(line)-1])
    return stopList

#获得停用词列表，用于对停用词过滤
stopList = getStopWords()
print(stopList)

#保存词频的词典
spamDict={}
normDict={}
testDict={}
#保存每封邮件中出现的词
wordsList=[]
wordsDict={}
#保存预测结果,key为文件名，值为预测类别
testResult={}

#获得每封邮件中的词，建立字典
def get_word_list(content, wordsList, stopList):
    res_list = list(jieba.cut(content))
    for word_i in res_list:
        if word_i not in stopList and word_i.strip()!='' and word_i != None:
            if word_i not in wordsList:
                wordsList.append(word_i)
                
#把词添加到字典
def add_to_dict(wordsList, wordsDict):
    for item in wordsList:
        if item in wordsDict.keys():
            wordsDict[item] += 1 
        else:
            wordsDict.setdefault(item, 1)
            
#正常邮件词频
for fileName in normFileList:
    wordsList.clear()
    for line in open(r"F:\NLP_learning\BayesSpam-master\data\normal"+'\\'+fileName):
                     rule=re.compile(r"[^\u4e00-\u9fa5]")
                     line = rule.sub("", line)
                     get_word_list(line, wordsList, stopList)
    add_to_dict(wordsList, wordsDict)
    
normDict=wordsDict.copy()
normDict

#垃圾邮件词频
wordsDict.clear()
for fileName in spamFileList:
    wordsList.clear()
    for line in open(r"F:\NLP_learning\BayesSpam-master\data\spam"+'\\'+fileName):
                     rule=re.compile(r"[^\u4e00-\u9fa5]")
                     line = rule.sub("", line)
                     get_word_list(line, wordsList, stopList)
    add_to_dict(wordsList, wordsDict)
    
spamDict=wordsDict.copy()
spamDict
```

```python
#通过计算每个文件中p(w|s)和p(w|n)来得到对分类影响最大的15个词
def get_test_words(testDict,spamDict,normDict,normFilelen,spamFilelen):
    wordProbListSpam = {}
    wordProbListNorm = {}
    
    for word, num in testDict.items():
        if word in spamDict.keys() and word in normDict.keys():
            #该文件中包含词个数
            #P(w|s) ---词w 在垃圾邮件中出现的频次
            pw_s = spamDict[word] / spamFilelen
            #P(w|n) ---词w 在正常邮件中出现的频次
            pw_n = normDict[word] / normFilelen
            
#             ps_w = pw_s / (pw_s + pw_n)
            
            wordProbListSpam.setdefault(word, pw_s)
            wordProbListNorm.setdefault(word, pw_n)
            
        if word in spamDict.keys() and word not in normDict.keys():
            pw_s = spamDict[word] / spamFilelen
            pw_n = 0.1 / normFilelen  #平滑
#             ps_w = pw_s / (pw_s + pw_n)
            
            wordProbListSpam.setdefault(word, pw_s)
            wordProbListNorm.setdefault(word, pw_n)
            
        if word not in spamDict.keys() and word in normDict.keys():
            pw_s=0.1 / spamFilelen
            pw_n=normDict[word]/normFilelen
#             ps_w=pw_s/(pw_s+pw_n) 
            
            wordProbListSpam.setdefault(word, pw_s)
            wordProbListNorm.setdefault(word, pw_n)
            
        if word not in spamDict.keys() and word not in normDict.keys():
            pw_s=0.1 / spamFilelen
            pw_n=0.1 / normFilelen
            
            wordProbListSpam.setdefault(word, pw_s)
            wordProbListNorm.setdefault(word, pw_n)
            
#     sorted(wordProbListSpam.items(),key=lambda d:d[1],reverse=True)[0:15]
#     sorted(wordProbListNorm.items(),key=lambda d:d[1],reverse=True)[0:15]
    
    return (sorted(wordProbListSpam.items(),key=lambda d:d[1],reverse=True)[0:15], sorted(wordProbListNorm.items(),key=lambda d:d[1],reverse=True)[0:15])
```

```python
#计算贝叶斯公式
def calBayes(wordProbListSpam, wordProbListNorm,spamdict,normdict):
    ps_w=1
    ps_n=1
    pw_s = 1
    pw_n = 1
    
    for word,prob in wordProbListSpam:
#         print(word+"/"+str(prob))
        pw_s*=(prob)
        
    for word,prob in wordProbListNorm:
#         print(word+"/"+str(prob))
        pw_n*=(prob)  
    
    p=(pw_s)/(pw_n)
    return p  
```

```python
def calAccuracy(testResult):
    rightCount=0
    errorCount=0
    for name ,catagory in testResult.items():
        if (int(name)<1000 and catagory==0) or(int(name)>1000 and catagory==1):
            rightCount+=1
        else:
            errorCount+=1
            
    return rightCount/(rightCount+errorCount)
```

```python
#测试邮件
for fileName in testFileList:
    testDict.clear()
    wordsDict.clear()
    wordsList.clear()
    for line in open(r"F:\NLP_learning\BayesSpam-master\data\test"+'\\'+fileName):
        rule=re.compile(r"[^\u4e00-\u9fa5]")
        line=rule.sub("",line)
        get_word_list(line,wordsList,stopList)
        
    add_to_dict(wordsList, wordsDict)
    testDict=wordsDict.copy()
    
    #通过计算每个文件中p(s|w)来得到对分类影响最大的15个词
    wordProbListSpam,wordProbListNorm=get_test_words(testDict, spamDict,normDict,normFilelen,spamFilelen)
    #对每封邮件得到的15个词计算贝叶斯概率  
    p=calBayes(wordProbListSpam, wordProbListNorm, spamDict, normDict)
    if(p>0.5):
        testResult.setdefault(fileName,1)
    else:
        testResult.setdefault(fileName,0)
```

```python
testAccuracy=calAccuracy(testResult)

for i,ic in testResult.items():

    print(i+"/"+str(ic))

print(testAccuracy)  
```
